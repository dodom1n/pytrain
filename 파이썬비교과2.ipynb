- 토큰화
- 토큰 : 자연어에서 의미가 있으면서 가장 작은 단위
- 토큰화 -> 벡터화

```python
okt = Okt()
kkma = Kkma( )

data1 = abstract_list

# 키워드 추출
data2 = []

for i in data1:
    data2.append(okt.nouns(i))
```

```python
print(data1[1])
print(data2[1])
```

```python
# 추출한 명사를 한 덩어리로 합체
data3 = sum(data2, [])

# 각 단어(명사)의 빈도 count
data4 = Counter(data3)

# data4 에서 최빈 단어 100개 추출
data5 = data4.most_common(100)
```

```python
len(data5[11])
```

```python
# 불용어 (Stop Words)
# 글자수 기준 불용어 제거

data6 = []
for i in data5:
    if len(i[0]) >= 2 and len(i[0]) <= 10:
        data6.append(i)

data6
```

```python
# 지정단어 삭제
# sword = open('불용어사전.txt').read()
sword = ['파이썬', '의미', '학습']

data7 = [i for i in data6 if i[0] not in sword]
data7
```

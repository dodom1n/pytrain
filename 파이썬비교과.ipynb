### print(변수 + 문자열 조합해서 출력하기)

```python
변수_1 = 5
변수_2 = 3

#방법1 : 변수와 문자열을 콤마로 나누어주기
print(변수_1, '+', 변수_2, '=', 변수_1 + 변수_2)

#방법2 : %를 이용하기
print('%s + %s = %s' %(변수_1, 변수_2, 변수_1 + 변수_2))

#방법3 : format 함수를 이용하기
문자열_샘플 = '{} + {} = {}'
print(문자열_샘플.format(변수_1, 변수_2, 변수_1 + 변수_2))

#방법4 : f와 {}를 이용하기
print(f'{변수_1} + {변수_2} = {변수_1 + 변수_2}')
```

## 리스트용 함수

```python
# 데이터 추가하기
# append() / insert()
list3 = ['첫째','둘째','셋째']
list3.append("append로 추가한 내용")
list3
```

```python
list3.insert(1, 'insert로 추가한 내용')
list3
```

```python
# 리스트에서 데이터 삭제하기
# pop() : append의 반대 -> 맨 뒤에 있는 데이터 삭제
list3.pop()
list3
```

```python
# 인덱스 기준 삭제
del list3[1]
list3
```

```python
# 값 기준 삭제
list3.remove('둘째')
```
